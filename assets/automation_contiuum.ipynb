```python
import matplotlib.pyplot as plt


with plt.xkcd():
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 6))

    # Data
    x = [0.2, 0.5, 0.8]  # Control levels: increasing control should decrease as efficiency increases
    y = [0.2, 0.6, 0.9]  # Efficiency levels
    labels = ['Manual', 'Semi-Automatic', 'Fully Automatic']
    colors = ['#4a90e2', '#9b59b6', '#50c878']
    # change all colors to black
    colors = ['#000000', '#000000', '#000000'] 


    # Adjusting semi-automatic point to lie on the line
    x[1] = 0.5
    y[1] = 0.55

    # Plot points
    ax.scatter(x, y, s=100, c=colors, zorder=3)

    # Add labels to points
    for i, label in enumerate(labels):
        ax.annotate(label, (x[i], y[i]), xytext=(0, 10), 
                    textcoords='offset points', ha='center', va='bottom',
                    fontsize=12, fontweight='bold')

    # Plot a straight line connecting the points
    ax.plot([0.2, 0.8], [0.2, 0.9], 'k--', alpha=0.5, linewidth=2, zorder=2)

    # Set labels and title
    ax.set_xlabel('Control', fontsize=14, fontweight='bold')
    ax.set_ylabel('Efficiency', fontsize=14, fontweight='bold')
    ax.set_title('Automation Continuum', fontsize=16, fontweight='bold')

    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Set axis limits and remove numbers from axes
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])

    # Invert x-axis so control decreases as efficiency increases
    ax.invert_xaxis()

    # Adjust layout and display the plot
    plt.tight_layout()

    # Save the simplified chart
    chart_path= "automation_continuum.png"
    plt.savefig(chart_path, bbox_inches='tight')
    plt.show()

```
